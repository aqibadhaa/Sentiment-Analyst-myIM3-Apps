```python
import google_play_scraper
from google_play_scraper import reviews_all, Sort
import pandas as pd

# ID aplikasi myIM3 di Google Play Store
app_id = 'com.pure.indosat.care'

# Mengambil semua ulasan
reviews = reviews_all(
    app_id,                   # ID aplikasi
    lang='id',                 # Bahasa ulasan (id = Bahasa Indonesia)
    country='id',              # Negara (id = Indonesia)
    sort=Sort.MOST_RELEVANT,   # Urutan ulasan
    count=50000                 # Maksimal 50000 ulasan
)

# Mengubah hasil scraping menjadi DataFrame
reviews_df = pd.DataFrame(reviews)

# Menyimpan DataFrame ke file CSV
reviews_df.to_csv('myim3_ulasan.csv', index=False)

print("Data ulasan berhasil disimpan ke dalam file 'myim3_ulasan.csv'")
```
